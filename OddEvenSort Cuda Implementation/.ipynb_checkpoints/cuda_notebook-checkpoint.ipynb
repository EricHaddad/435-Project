Groupby matrix size argument, available in the metadata (tk.metadata) to organize our performance data into 3 groupings.

```python
tk.metadata
```

```python
gb = tk.groupby("num_vals")
```

Compose these 3 groups column-wise to get a holistic view of the data

```python
ctk = th.Thicket.concat_thickets(
    thickets=list(gb.values()),
    headers=list(gb.keys()),
    axis="columns",
    metadata_key="num_threads"
)
```

```python
ctk.dataframe
```

Perform pandas operations to get the table in a desirable format for plotting

```python
# Drop unwanted columns
ctk.dataframe = ctk.dataframe.reset_index().drop([("node", ""), (65536, "nid"), (1048576, "nid"), (16777216, "nid")], axis=1)
#ctk.dataframe = ctk.dataframe.reset_index().drop([("node", ""), (128, "Total time"), (1024, "Total time"), (8192, "Total time")], axis=1)
# Rename index to remove tuple
ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])
# Drop "time" label as it is redundant
#ctk.dataframe.columns = ctk.dataframe.columns.droplevel(1)
```

```python
ctk.dataframe
```

```python
# Define common variables
function_names = set(ctk.dataframe.index.get_level_values("name"))
print(function_names)
threads = sorted(list(set(ctk.dataframe.index.get_level_values("num_threads"))))
print(threads)
matrix_sizes = sorted(ctk.dataframe.columns.tolist())
```

# Plotting

### Option 1
Use `pandas.DataFrame.plot` to automatically interface with matplotlib

```python
def plot_pandas(df, title, xlabel, ylabel, loglog):
    plot = df.plot(kind="line", 
        marker='o', 
        title=title, 
        xlabel=xlabel,
        ylabel=ylabel, 
        grid=True,
        loglog=loglog
    )
    return plot
```

### Option 2
Use `matplotlib.pyplot` directly for more fine-grained control. Specifically for our case we can change the x axis scaling from base 10 to base 2.

```python
def plot_mpl(df, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    
    ax.plot(
        df[0],
        marker='o',
    )
    ax.plot(
        df[1],
        marker='o',
    )
    
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(["cudaMemcpy_device_to_host", "cudaMemcpy_host_to_device"])
    plt.grid(True)
    plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(threads)
    plt.show()
    return fig, ax
```

```python
for name in function_names:
    plot_pandas(
        ctk.dataframe.loc[name][128],
        title=name + " for matrix size 128",
        xlabel="Threads",
        ylabel="Time (seconds)",
        loglog=False    
    )
```
